```python
data = pd.read_csv("/content/drive/MyDrive/물공모전/서산_토성 채워조.csv", encoding='euc-kr')

from datetime import datetime

def date_range(start, end):
    start = datetime.strptime(start, "%Y-%m-%d")
    end = datetime.strptime(end, "%Y-%m-%d")
    dates = [date.strftime("%Y-%m-%d") for date in pd.date_range(start, periods=(end-start).days+1)]
    return dates

dates = date_range("2021-01-01", "2021-01-09")
print(dates)

#날짜 리스트 만드는 함수 제작
def make_date_list(start, end):
  year = ["2018-", "2019-", "2020-", "2021-", "2022-"]
  date_list = []
  for y in year:
    start_date = y + start
    end_date = y + end
    date_tmp = date_range(start_date, end_date)
    date_list = date_list + date_tmp
  return date_list
```

```python
koguma_g1 = make_date_list('05-15', '06-10')


koguma1_df= data[data['일시'].isin(koguma_g1)]
koguma1_df

koguma1_미사질식양토 = koguma1_df.copy().reset_index()
koguma1_미사질식양토['토양'] = '미사질식양토'
koguma1_미사질식양토['생육단계'] = 'G1'
koguma1_미사질식양토['1회관계량'] = 18.51
koguma1_미사질식양토
```

```python
def make_df(crop, data, soil, g_num, g_list, water_list):

  global df_i
  global df_fin

  for i in range(len(g_num)):
    start = g_list[i*2]
    end = g_list[i*2 + 1]
    tmp_date = make_date_list(start, end)
    df= data[data['일시'].isin(tmp_date)]

    for k in range(len(soil)):
      df_s = df.copy().reset_index()
      df_s['토양'] = soil[k]
      df_s['생육단계'] = g_num[i]
      df_s['1회관계량'] = water_list[k][i]
      df_s['작물'] = crop

      if soil[k] == '미사질식양토':
        df_i = df_s
      else:
        df_i = df_i.append(df_s, ignore_index=True)

    if i == 0:
      df_fin = df_i
    else:
      df_fin = df_fin.append(df_i, ignore_index=True)

  return df_fin
```
